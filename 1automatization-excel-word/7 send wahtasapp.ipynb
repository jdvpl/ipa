```python
import webbrowser
import pyautogui
from time import sleep
```

```python
def send_whatsapp_message(phone_number, messages):
    webbrowser.open('https://web.whatsapp.com/send?phone=' + phone_number)
    sleep(8)  # Ajusta este tiempo según sea necesario para que la página se cargue completamente
    
    for i in range(len(messages)):
        message = messages[i]
        pyautogui.typewrite(message)
        pyautogui.press('enter')
        sleep(2)
```
