```python
import tensorflow as tf
import tensorflow_datasets as tfData
import matplotlib.pyplot as plt
```

```python
data, metaData = tfData.load('fashion_mnist', as_supervised=True, with_info=True)
```

```python
metaData
```

```python
trainingData=data['train']
testData=data['test']
```

```python
nameClasses=metaData.features['label'].names
```

```python
nameClasses
```

```python
def normalizeImage(images, tags):
    images=tf.cast(images,tf.float32)
    images/=255
    return images, tags
trainingData=trainingData.map(normalizeImage)
testData=testData.map(normalizeImage)
```

```python
trainingData=trainingData.cache()
testData=testData.cache()
```

```python
for image, tag, in trainingData.take(1):
    break
image=image.numpy().reshape((28,28))
plt.figure()
plt.imshow(image,cmap=plt.cm.binary)
plt.colorbar()
plt.grid(False)
plt.show()
```

```python
plt.figure(figsize=(10,10))
for i,(image, tag) in enumerate(trainingData.take(25)):
    image=image.numpy().reshape((28,28))
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(image,cmap=plt.cm.binary)
    plt.xlabel(nameClasses[tag])
plt.show()
```

```python
model = tf.keras.Sequential([
    tf.keras.Input(shape=(28, 28, 1)),  # Definir la forma de entrada explícitamente
    tf.keras.layers.Flatten(),  # Aplanar la entrada
    tf.keras.layers.Dense(50, activation=tf.nn.relu),
    tf.keras.layers.Dense(50, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
```
